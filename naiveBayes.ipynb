```python
from datasets import load_dataset

ds = load_dataset("google-research-datasets/poem_sentiment")
```

```python
for split, subset in ds.items():
    print(split, len(subset))
```

```python
# Prepare data for sklearn
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn import metrics

# inspect an example
print(ds['train'][0])

# dataset uses 'verse_text' field
train_texts = [x['verse_text'] for x in ds['train']]
train_labels = [x['label'] for x in ds['train']]

test_texts = [x['verse_text'] for x in ds['test']]
test_labels = [x['label'] for x in ds['test']]

# build pipeline: bag-of-words -> tfidf -> naive bayes
clf = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', MultinomialNB()),
])

# train
clf.fit(train_texts, train_labels)

# predict on test
predicted = clf.predict(test_texts)
acc = metrics.accuracy_score(test_labels, predicted)
print(f"Test accuracy: {acc:.4f}")
print(metrics.classification_report(test_labels, predicted))
```
